```python
pd.options.display.max_rows = 1000
data3 = pd.read_csv ("/Users/giuliocassis/desktop/sqlproject/dayofweek.csv")
data3.head()
```

```python
#Making list within list of number of tests completed on each day of the week, seperated by months. 
#For example first list within the list has 7 values, first value representing tests completed on Sunday in January, 2nd value representing total tests completed on Monday in January. 
#2nd list within list also has 7 values, same idea as first list,but represents Feburary. 

total = [[] for month in range(12)]
i = 0 
while i < len(data3.Total_Tests): 
    if data3.The_Month[i] == 1:
        total[0] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 2:
        total[1] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 3:
        total[2] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 4:
        total[3] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 5:
        total[4] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 6:
        total[5] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 7:
        total[6] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 8:
        total[7] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 9:
        total[8] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 10:
        total[9] += [data3.Total_Tests[i]]
    elif data3.The_Month[i] == 11:
        total[10] += [data3.Total_Tests[i]]
    else: 
        total[11] += [data3.Total_Tests[i]]
    i += 1

simplified=[]
for i in range(7):
    simplified.append(data3.Day_of_Week[i][:3])
```

```python
#Making an overall bar chart representing number of complete tests by each day of the week, looking at the year as a whole.


#converting data to numpy ("variablename" = "dataframename".values)
df = data3.values
df

#Create 7 lists representing each day 
#within each list values which represent row number 
day = []
rows = []
for i in range (7):
    while i < len(df): 
        day += [i] 
        i += 7 
    rows += [day]
    day = []


#1D array for each day 
sunday = df[rows[0], 3]
monday = df[rows[1], 3]
tuesday = df[rows[2], 3]
wednesday = df[rows[3], 3]
thursday = df[rows[4], 3]
friday = df[rows[5], 3]
saturday = df[rows[6], 3]

sunday2 = np.array(sunday)
monday2 = np.array(monday)
tuesday2 = np.array(tuesday)
wednesday2 = np.array(wednesday)
thursday2 = np.array(thursday)
friday2 = np.array(friday)
saturday2 = np.array(saturday)


#combining individual arrays into a 2D array using np.vstack
total = np.vstack((sunday2, monday2, tuesday2, wednesday2, thursday2, friday2, saturday2))
#summing up values in each individual array to get total number of tests conducted each day of week 
#COMPLETE
final = np.sum(total/12, axis = 1)



#Generating Bar-Graph 
plt.figure(figsize = (4,5))
plt.bar(simplified, final, width = 0.7, edgecolor = "black", linewidth = "2" )
plt.title ("Total Tests Completed by Day of Week ", loc = "center")
axes = plt.gca()
axes.set_ylim([0,3000])
plt.show()
```
