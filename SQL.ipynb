```python
#Looking into demographics of Dognition Users by Country.

%%sql 
SELECT users.country as Country, 
COUNT(users.user_guid) as Total_Users, 
AVG(users.max_dogs) as Avg_Num_Dogs, 
ROUND(COUNT(users.user_guid) * AVG(users.max_dogs),0) as Total_Dogs
FROM users JOIN dogs ON users.user_guid = dogs.user_guid 
WHERE Country NOT IN ("None", "N/A") and users.max_dogs >= 1
GROUP BY Country
ORDER BY Total_Dogs DESC
LIMIT 15;
```

```python
#Uploading data into csv file

demographic= %%sql SELECT users.country as Country, COUNT(users.user_guid) as Total_Users, AVG(users.max_dogs) as Avg_Num_Dogs, (COUNT(users.user_guid) * AVG(users.max_dogs)) as Total_Dogs FROM users JOIN dogs ON users.user_guid = dogs.user_guid WHERE Country NOT IN ("None", "N/A") and users.max_dogs >= 1 GROUP BY Country ORDER BY Total_Dogs DESC;
demographic.csv("demographics.csv")
```

```python
#Now looking into distribution of Dognition Users by States in the US.

%%sql 
SELECT users.state,
CASE WHEN users.state IN ("AZ","CO", "ID", "MT", "NV", "NM", "UT", "WY","AK", "CA", "HI", "OR", "WA") THEN "West"
WHEN users.state IN ("ME","NH","MA","RI","CT","NY","NJ","PA","VT") THEN "North-East"
WHEN users.state IN ("IL","IN","IA","KS","MI","MN","MO","NE","ND","SD","OH","WI") THEN "Mid-West"
ELSE "South"
END AS State_Classification, 
COUNT(users.user_guid) as Total_Users, 
AVG(users.max_dogs) as Avg_Num_Dogs, 
ROUND(COUNT(users.user_guid) * AVG(users.max_dogs), 0) as Total_Dogs
FROM users JOIN dogs ON users.user_guid = dogs.user_guid 
WHERE Country = "US"  and users.max_dogs >= 1
GROUP BY users.state
ORDER BY Total_Dogs DESC; 
```

```python
#Looking into whether Number of Complete Tests varies by Day of the Week.

%%sql 
SELECT MONTH(created_at) as The_Month, DAYOFWEEK(created_at) as Day_Week, 
CASE WHEN DAYOFWEEK(created_at)= 1 THEN "Sunday"
WHEN DAYOFWEEK(created_at) = 2 THEN "Monday" 
WHEN DAYOFWEEK(created_at) = 3 THEN "Tuesday" 
WHEN DAYOFWEEK(created_at) = 4 THEN "Wednesday"
WHEN DAYOFWEEK(created_at) = 5 THEN "Thursday" 
WHEN DAYOFWEEK(created_at) = 6 THEN "Friday" 
ELSE "Saturday"
END AS Day_of_Week,
COUNT(created_at) as Total_Tests
FROM complete_tests 
GROUP BY The_Month, Day_Week, Day_of_Week; 
```

```python
#Uploading data into csv file

dayofweek = %%sql SELECT MONTH(created_at) as The_Month, DAYOFWEEK(created_at) as Day_Week, CASE WHEN DAYOFWEEK(created_at)= 1 THEN "Sunday" WHEN DAYOFWEEK(created_at) = 2 THEN "Monday" WHEN DAYOFWEEK(created_at) = 3 THEN "Tuesday" WHEN DAYOFWEEK(created_at) = 4 THEN "Wednesday" WHEN DAYOFWEEK(created_at) = 5 THEN "Thursday" WHEN DAYOFWEEK(created_at) = 6 THEN "Friday" ELSE "Saturday" END AS Day_of_Week, COUNT(created_at) as Total_Tests FROM complete_tests GROUP BY The_Month, Day_Week, Day_of_Week;
dayofweek.csv("dayofweek.csv")
```

```python
#Checking for possible relationship between dog-type (dogs.dimension) and number of complete tests for each test.

%%sql 
SELECT complete_tests.test_name, COUNT(complete_tests.created_at),
COUNT(CASE WHEN dogs.dimension IS NULL THEN complete_tests.created_at END) AS "N/A", 
COUNT(CASE WHEN dogs.dimension = "ace" THEN complete_tests.created_at END) AS Ace, 
COUNT(CASE WHEN dogs.dimension = "charmer" THEN complete_tests.created_at END) AS Charmer, 
COUNT(CASE WHEN dogs.dimension = "einstein" THEN complete_tests.created_at END) AS Einstein, 
COUNT(CASE WHEN dogs.dimension = "expert" THEN complete_tests.created_at END) AS Expert, 
COUNT(CASE WHEN dogs.dimension = "maverick" THEN complete_tests.created_at END) AS Maverick, 
COUNT(CASE WHEN dogs.dimension = "protodog" THEN complete_tests.created_at END) AS Protodog, 
COUNT(CASE WHEN dogs.dimension = "renaissance-dog" THEN complete_tests.created_at END) AS Renaissance,
COUNT(CASE WHEN dogs.dimension = "socialite" THEN complete_tests.created_at END) AS Socialiate, 
COUNT(CASE WHEN dogs.dimension = "stargazer" THEN complete_tests.created_at END) AS Stargazer 
FROM complete_tests JOIN dogs ON complete_tests.dog_guid = dogs.dog_guid
WHERE complete_tests.created_at IS NOT NULL and (dogs.exclude != 1 or dogs.exclude IS NOT NULL)
GROUP BY complete_tests.test_name
ORDER BY COUNT(complete_tests.created_at) DESC;
```

```python
#Perhaps there is a relationship between AVG Days Between Tests (dogs.mean_iti_days) and Number of Tests Completed. For instance if a user has their dog take tests at large time intervals (high dogs.mean_iti_days) , then number of tests completed may be low because it shows they are not that commited to having their dogs complete the games.

%%sql
SELECT users.membership_type, dogs.mean_iti_days AS "AVG Days Between Tests", dogs.total_tests_completed AS "Total Tests Completed"
FROM users JOIN dogs on users.user_guid = dogs.user_guid
WHERE (dogs.exclude != 1 or dogs.exclude IS NOT NULL) and dogs.mean_iti_days != "NaN"
ORDER BY (dogs.mean_iti_days) DESC
LIMIT 100;
```

```python
#Uploading data into csv file

data4 = %%sql SELECT users.membership_type, dogs.mean_iti_days AS "AVG Days Between Tests", dogs.total_tests_completed AS "AVG Tests Completed" FROM users JOIN dogs on users.user_guid = dogs.user_guid WHERE (dogs.exclude != 1 or dogs.exclude IS NOT NULL) and dogs.mean_iti_days != "NaN" ORDER BY (dogs.mean_iti_days) DESC;
data4.csv("data4.csv")
```
